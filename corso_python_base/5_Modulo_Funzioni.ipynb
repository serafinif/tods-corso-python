## ðŸ§© Parametri e argomenti
Python supporta diversi tipi di parametri:

### 1ï¸âƒ£ Parametri posizionali e keyword

Le funzioni possono ricevere uno o piÃ¹ **parametri** (input).
Quando vengono chiamate, i valori passati si chiamano argomenti.

**ðŸ”¹ Parametri posizionali**

Gli argomenti vengono associati in base allâ€™ordine.

**ðŸ”¹ Parametri keyword (per nome)**

Ãˆ possibile specificare gli argomenti esplicitamente per nome:

````python
def descrivi_persona(nome, eta):
    print(f"{nome} ha {eta} anni.")

descrivi_persona('Alice', 25)           # posizionali
descrivi_persona(eta=30, nome='Bob')    # keyword (ordine libero)
````

> ðŸ’¡ I parametri posizionali vengono assegnati in ordine;
>
> i keyword argument rendono il codice piÃ¹ leggibile e meno soggetto a errori.

### 2ï¸âƒ£ Valori di default

Puoi assegnare un **valore predefinito** a uno o piÃ¹ parametri.

Se lâ€™argomento non viene passato, viene usato il valore di default.

````python
def saluto(nome='ospite'):
    print(f'Ciao {nome}!')

saluto()
saluto('Luca')
````

> ðŸ’¡ I valori di default devono sempre venire dopo i parametri obbligatori.

```python
def esempio(a, b=10):  # âœ… corretto
    ...

    ...
```

### 3ï¸âƒ£ Argomenti variabili: *args e **kwargs

A volte non si conosce in anticipo **quanti argomenti** riceverÃ  una funzione.

Python permette di gestire questa flessibilitÃ  con due sintassi speciali:

| Sintassi   | Nome                 | Significato                                       |
| ---------- | -------------------- | ------------------------------------------------- |
| `*args`    | positional arguments | raccoglie argomenti posizionali in una **tupla**  |
| `**kwargs` | keyword arguments    | raccoglie argomenti nominati in un **dizionario** |

**ðŸ”¹ Esempio con `*args`**

````python
def somma_tutti(*args):
    print("Hai passato:", args)
    return sum(args)

print(somma_tutti(2, 4, 6, 8, 9, 56, 34))  
````

````python
print(somma_tutti(5,10,15,20))
````

**ðŸ”¹ Esempio con `**kwargs`**

````python
def descrivi_persona(**kwargs):
    print("Dati ricevuti:", kwargs)
    for chiave, valore in kwargs.items():
        print(f"{chiave:13}: {valore}")

descrivi_persona(nome="Ada", eta=36, professione="scienziata")
````

ðŸ‘‰ In questo caso dentro la funzione avrai:

```python
kwargs == {'nome': 'Ada', 'etÃ ': 36, 'professione': 'scienziata'}
```

> ðŸ’¡ *args e **kwargs possono anche coesistere:
>
>   ```python
>   def funzione(a, *args, **kwargs):
>       ...
>   ```


````python
def funzione(a, b, *args, **kwargs):
    print("a e b:", a, b)
    print("args:", args)       # tupla di argomenti posizionali extra
    print("kwargs:", kwargs)   # dizionario di argomenti keyword extra

funzione(1, 2, 3, 4, x=10, y=20)
````

````python
def funzione(a, b, **kwargs):
    print("a e b:", a, b)
    print("kwargs:", kwargs)   # dizionario di argomenti keyword extra

funzione(1, 2, x=10, y=20)
````

````python
def funzione(a,b,**kwargs):
    print("a e b:", a, b)
    print("kwargs:", kwargs)   # dizionario di argomenti keyword extra

funzione(1, 2,x=10, y=20)
````

**ðŸ’¡ Regola dâ€™ordine:**

```python
    parametri normali â†’ *args â†’ **kwargs
```

perchÃ© `**kwargs` deve sempre venire **per ultimo**.

**ðŸ”¹ Espansione di dizionari con**

Puoi anche **passare un dizionario** a una funzione che accetta `**kwargs` â€” usando di nuovo `**` per â€œsrotolareâ€ le coppie chiave/valore come argomenti keyword:

````python
def mostra_dati(**kwargs):
    print(kwargs)

dati = {"nome": "Ada", "etÃ ": 36}
mostra_dati(**dati)  # Espande il dizionario in argomenti keyword
````

ðŸ‘‰ Ãˆ come scrivere:

````python
mostra_dati(nome="Ada", etÃ =36)
````

````python
def mostra_dati(*args):
    print(args)

dati = ("Ada", 36)
mostra_dati(*dati)  # Espande la tupla in argomenti posizionali
````

**ðŸ§© In sintesi**

| Simbolo    | Tipo               | Raccoglie in | Esempio              |
| ---------- | ------------------ | ------------ | -------------------- |
| `*args`    | Posizionali        | Tupla        | `(1, 2, 3)`          |
| `**kwargs` | Nominati (keyword) | Dizionario   | `{'x': 10, 'y': 20}` |


## ðŸŒ Scope delle variabili (locale vs globale)

Ogni variabile ha un ambito (scope) che definisce dove Ã¨ visibile.

Le **variabili locali** vivono solo dentro la funzione.
Le **variabili globali** sono accessibili ovunque, ma **non modificate** senza `global`.

| Tipo        | Dove vive              | AccessibilitÃ     |
| ----------- | ---------------------- | ---------------- |
| **Locale**  | dentro una funzione    | visibile solo lÃ¬ |
| **Globale** | fuori da ogni funzione | visibile ovunque |


**ðŸ”¹ Esempio:**

````python
x = 10  # variabile globale

def f():
    x = 5  # locale
    print('Dentro la funzione:', x)

f()
print('Fuori dalla funzione:', x)
````

**ðŸ”¹ Uso di global**

Per modificare una variabile globale dentro una funzione:

````python
contatore = 0

def incrementa():
    global contatore
    contatore += 1

incrementa()
print(contatore)  # 1
````

> âš ï¸ In generale, evita di modificare variabili globali: riduce la chiarezza del codice.

## ðŸ§¾ Documentazione con Docstring

Le **docstring** descrivono cosa fa una funzione, sono stringhe speciali che descrivono cosa fa una funzione. 

Vanno scritte **subito dopo la definizione** tra triple virgolette (`"""`).

Possono essere lette con `help()` o `.__doc__`.

````python
def area_rettangolo(base: float, altezza: float) -> float:
    '''Calcola l'area di un rettangolo.
    Args:
        base (float): base del rettangolo
        altezza (float): altezza del rettangolo
    Returns:
        float: area calcolata'''
    return base * altezza

help(area_rettangolo)
print('Area =', area_rettangolo(5, 3))
````

````python
# oppure
print(area_rettangolo.__doc__)
````

> ðŸ’¡ Le docstring migliorano la leggibilitÃ  e manutenibilitÃ  del codice.

## ðŸ§® Introduzione ai Type Hints

I **Type Hints** indicano il tipo previsto dei parametri e del valore di ritorno.  
Possono essere usati insieme ai **valori di default** per rendere la funzione ancora piÃ¹ chiara e sicura.

**ðŸ”¹ Sintassi generale**

```python
def nome_funzione(parametro: tipo = valore_default) -> tipo_ritorno:
    ...
```

- `parametro: tipo` â†’ specifica il tipo del parametro
- `= valore_default` â†’ assegna un valore di default se non viene passato
- `-> tipo_ritorno` â†’ indica il tipo del valore restituito

**ðŸ”¹ Esempio base**

````python
def saluta(nome: str = "ospite") -> None:
    print(f"Ciao, {nome}!")
saluta()
saluta("Luca")
saluta(42)  # Anche se non ha senso, Ã¨ permesso
saluta(None)
saluta((2,3))
````

**ðŸ“˜ Spiegazione:**

- `nome: str` â†’ il parametro nome Ã¨ una stringa
- `= "ospite"` â†’ se non passo nulla, userÃ  "ospite"
- `-> None` â†’ la funzione non restituisce nulla (solo stampa)

````python
def area_cerchio(raggio: float) -> float:
    '''Calcola l'area di un cerchio usando hint di tipo.'''
    import math
    return math.pi * raggio ** 2

print(area_cerchio(5.0))
````

````python
print(area_cerchio(5))
````

````python
print(area_cerchio("5.0"))
````

>ðŸ§  Python **non applica controlli di tipo a runtime**, ma i type hints aiutano strumenti come mypy o VS Code a segnalare errori prima dellâ€™esecuzione.

**ðŸ”¹ Esempio con piÃ¹ parametri**

````python
def calcola_sconto(prezzo: float, sconto: float = 0.1) -> float:
    """
    Calcola il prezzo scontato.
    
    Parametri:
        prezzo (float): prezzo originale
        sconto (float): percentuale di sconto (default 0.1 = 10%)
    Ritorna:
        float: prezzo scontato
    """
    return prezzo * (1 - sconto)
````

````python
print(calcola_sconto(100))        # Usa sconto 10% â†’ 90.0
print(calcola_sconto(100, 0.25))  # Usa sconto 25% â†’ 75.0
````

**ðŸ”¹ Esempio con piÃ¹ tipi (opzionale con Optional o Union)**

Puoi anche specificare **piÃ¹ tipi ammessi**, ad esempio con `Optional` o `Union`.

Il tipo `Optional` (dal modulo `typing`) serve per indicare che un parametro non Ã¨ mandatory.

**ðŸ”¹ Sintassi generale**

```python

def funzione(x: Optional[tipo1]) -> tipo_ritorno:
    ...
```

**ðŸ”¹ Esempio base**

````python
from typing import Optional

def descrivi_utente(nome: str, etÃ : Optional[int] = None) -> str:
    if etÃ  is not None:
        return f"{nome} ha {etÃ } anni."
    return f"{nome} non ha specificato l'etÃ ."
````

````python
print(descrivi_utente("Luca", 30))
print(descrivi_utente("Anna"))
````

ðŸ“˜ Optional[int] significa â€œint o Noneâ€.

Il tipo **`Union`** (dal modulo `typing`) serve per indicare che un valore puÃ² essere **di piÃ¹ tipi diversi**.

Ãˆ molto utile quando una funzione accetta piÃ¹ formati (es. `int` o `float`, oppure `str` o `None`).

**ðŸ”¹ Sintassi generale**

```python
from typing import Union

def funzione(x: Union[tipo1, tipo2, tipo3]) -> tipo_ritorno:
    ...
```

**ðŸ“˜ Significa:**

â€œx puÃ² essere di tipo `tipo1`, oppure `tipo2`, oppure `tipo3`.â€

**ðŸ”¹ Esempio base**

````python
from typing import Union

def quadrato(x: Union[int, float]) -> float:
    return x ** 2
````

````python
print(quadrato(3))     # âœ… int
print(quadrato(2.5))   # âœ… float
# print(quadrato("3"))  # âŒ il linter segnala: tipo non compatibile
````

**ðŸ”¹ `Union` nei parametri opzionali**

Quando un parametro puÃ² anche essere None, si usa una forma abbreviata con Optional:

````python
from typing import Optional

def descrivi_utente(nome: str, etÃ : Optional[int] = None) -> str:
    if etÃ  is not None:
        return f"{nome} ha {etÃ } anni."
    return f"{nome} non ha specificato l'etÃ ."
````

ðŸ’¡ `Optional[int]` Ã¨ equivalente a `Union[int, None]`.

**ðŸ”¹ `Union` nel valore di ritorno**

Puoi anche usarlo per dichiarare **piÃ¹ tipi di ritorno possibili**:

````python
def dividi(a: int, b: int) -> Union[float, str]:
    if b == 0:
        return "Errore: divisione per zero"
    return a / b
````

````python
print(dividi(10, 2))   # 5.0 (float)
print(dividi(10, 0))   # "Errore: divisione per zero" (str)
````

Dalla versione **3.10** puoi usare il simbolo `|` invece di Union (piÃ¹ leggibile):

````python
def quadrato(x: int | float) -> float:
    return x ** 2

def dividi(a: int, b: int) -> float | str:
    if b == 0:
        return "Errore"
    return a / b
````

````python
quadrato(4)
dividi(10, 2)
````

````python
quadrato(4.0)
dividi(10, 0)
dividi(10, 3)
````

**ðŸ’¡ In sintesi**

| Elemento | Descrizione | Esempio | Note |
|-----------|-------------|----------|------|
| **Parametro semplice** | Definisce un parametro obbligatorio | `x` | Deve essere passato sempre |
| **Tipo parametro** | Specifica il tipo atteso del parametro | `x: int` | Aiuta lâ€™editor e il linting |
| **Valore di default** | Imposta un valore automatico se non viene passato | `x=10` | Deve venire dopo i parametri obbligatori |
| **Entrambi insieme** | Combina tipo e valore di default | `x: int = 10` | Combinazione comune e leggibile |
| **Tipo di ritorno** | Specifica il tipo restituito | `-> float` | Si scrive dopo i parametri |
| **Parametri variabili** | Raccoglie piÃ¹ valori posizionali o keyword | `*args`, `**kwargs` | Rispettivamente tupla e dizionario |
| **Parametro opzionale** | PuÃ² essere un valore o `None` | `x: Optional[int] = None` | Equivale a `Union[int, None]` |
| **PiÃ¹ tipi ammessi (Union)** | Il parametro accetta piÃ¹ tipi diversi | `x: Union[int, float]` | Da Python 3.10: `x: int | float` |
| **Funzione senza ritorno** | Indica che non restituisce alcun valore | `-> None` | Tipico per funzioni che stampano o scrivono su file |
| **Funzione anonima (`lambda`)** | Piccola funzione in una riga | `lambda x: x ** 2` | Utile in `key=`, `map()`, `filter()` |
| **Type Hint su `args`/`kwargs`** | Specifica il tipo degli argomenti variabili | `*args: int`, `**kwargs: str` | Non obbligatorio, ma possibile |
| **Type Hint su ritorno multiplo** | Indica piÃ¹ tipi di output possibili | `-> Union[str, float]` | Da Python 3.10: `-> str | float` |


**ðŸ“˜ Esempio completo**

````python
from typing import Union, Optional

def calcola_prezzo(
    prezzo: float,
    sconto: float = 0.1,
    valuta: str = "â‚¬",
    note: Optional[str] = None
) -> Union[str, float]:
    """
    Calcola il prezzo scontato e lo restituisce formattato come stringa.

    Parametri:
        prezzo (float): prezzo originale
        sconto (float): percentuale di sconto (default 0.1 = 10%)
        valuta (str): simbolo di valuta (default "â‚¬")
        note (Optional[str]): eventuale descrizione

    Ritorna:
        Union[str, float]: prezzo finale (float) o stringa formattata
    """
    prezzo_finale = round(prezzo * (1 - sconto), 2)
    if note:
        return f"{prezzo_finale}{valuta} - {note}"
    return prezzo_finale
````

````python
print(calcola_prezzo(100))                       # 90.0
print(calcola_prezzo(100, 0.2, "$"))             # 80.0
print(calcola_prezzo(100, note="Offerta speciale"))  # "90.0â‚¬ - Offerta speciale"
````

## ðŸ§® Approfondimento â€“ Il modulo `typing`

Il modulo **`typing`** fornisce gli strumenti per indicare i **tipi di dato attesi** nei parametri e nei valori di ritorno delle funzioni.

ðŸ‘‰ Non cambia il comportamento del codice, ma:
- rende il codice **piÃ¹ leggibile**;
- aiuta lâ€™editor e il *linter* a segnalare errori;
- migliora lâ€™autocompletamento e la documentazione.

**ðŸ”¹ Importazione base**

```python
from typing import List, Tuple, Dict, Set, Union, Optional, Any, Callable, Literal
```

**ðŸ§± Tipi generici principali**

**1ï¸âƒ£ Liste, tuple, insiemi e dizionari**

| Tipo                 | Descrizione                           | Esempio           | Significato          |
| -------------------- | ------------------------------------- | ----------------- | -------------------- |
| `List[T]`            | Lista di elementi dello stesso tipo   | `List[int]`       | lista di interi      |
| `Tuple[T1, T2, ...]` | Tupla con tipi noti e lunghezza fissa | `Tuple[str, int]` | es. `("Ada", 36)`    |
| `Tuple[T, ...]`      | Tupla di lunghezza variabile          | `Tuple[int, ...]` | es. `(1,2,3,4)`      |
| `Set[T]`             | Insieme di elementi univoci           | `Set[str]`        | es. `{"a","b","c"}`  |
| `Dict[K, V]`         | Dizionario chiaveâ†’valore              | `Dict[str, int]`  | es. `{"a":1, "b":2}` |


**ðŸ“˜ Esempio:**

````python
from typing import List, Dict

def analizza(voti: List[int]) -> Dict[str, float]:
    return {
        "media": sum(voti) / len(voti),
        "max": max(voti),
        "min": min(voti)
    }
    
print(analizza([85, 90, 78, 92, 88]))
````

**2ï¸âƒ£ Any**

Indica che un valore puÃ² essere **di qualunque tipo**.
Utile in funzioni molto generiche (ma da usare con cautela).

````python
from typing import Any

def stampa_valore(x: Any) -> None:
    print(f"Tipo: {type(x)}, valore: {x}")

stampa_valore(42)
stampa_valore(3.14)
stampa_valore("Ciao")
````

>ðŸ’¡ Equivale a dire â€œnon specifico il tipo, puÃ² essere tuttoâ€.

**3ï¸âƒ£ Callable**

Serve per dichiarare un parametro che Ã¨ una funzione.

````python
from typing import Callable

def applica(funzione: Callable[[int, int], int], a: int, b: int) -> int:
    return funzione(a, b)

print(applica(lambda x, y: x + y, 3, 4))  # 7
print(applica(lambda x, y: x * y, 3, 4))  # 12
````

ðŸ“˜ Callable[[T1, T2], R] â†’ accetta tipi T1, T2 e restituisce R.

**4ï¸âƒ£ Literal**

Serve per limitare i valori ammessi a un insieme finito (es. costanti o modalitÃ ).

````python
from typing import Literal

def imposta_modalitÃ (modalitÃ : Literal["chiaro", "scuro"]) -> None:
    print(f"ModalitÃ  impostata su {modalitÃ }")
    
imposta_modalitÃ ("chiaro")  # âœ…
imposta_modalitÃ ("scuro")   # âœ…
#imposta_modalitÃ ("auto")    # âŒ il linter segnala: valore non valido ma non genera errore a runtime
````

**5ï¸âƒ£ TypeAlias (alias di tipo)**

Serve per definire un nome leggibile per un tipo complesso.

````python
from typing import List, Tuple, TypeAlias

Coppia = Tuple[str, int]
Rubrica: TypeAlias = List[Coppia]

def stampa_rubrica(r: Rubrica) -> None:
    for nome, tel in r:
        print(f"{nome}: {tel}")
````

````python
Rubrica = [("Alice", 12345), ("Bob", 67890)]
stampa_rubrica(Rubrica)
````

````python
persona_1: Persona = {
    "nome": "Luca",
    "etÃ ": 28,
    "email": "luca@dominio.com"
}
descrivi(persona_1)
````

## âš–ï¸ Funzioni pure vs con effetti collaterali

**ðŸ”¹ Funzioni pure**

Una **funzione pura**:

- restituisce **sempre lo stesso output** per lo stesso input;
- **non modifica** variabili o stati esterni.

````python
# Funzione pura
def quadrato(x):
    return x * x
````

âž¡ï¸ Ãˆ prevedibile, testabile e sicura.

**ðŸ”¹ Funzioni con effetti collaterali**

Una funzione che **interagisce con lâ€™esterno** (stampa, scrive su file, modifica variabili globali) ha **effetti collaterali**.

````python
# Effetto collaterale: modifica una lista globale
dati = []
def aggiungi_elemento(e):
    dati.append(e)
a=4
print(quadrato(a)) # non mi modifica a
aggiungi_elemento('nuovo') # mi modifica dati (lista globale)
print('Lista globale modificata:', dati)
````

âž¡ï¸ PuÃ² essere utile, ma riduce la prevedibilitÃ  del codice.

**ðŸ’¡ In sintesi**

| Tipo                           | Caratteristiche                                  | Esempi                           |
| ------------------------------ | ------------------------------------------------ | -------------------------------- |
| âœ… **Funzione pura**            | Deterministica, indipendente dallo stato esterno | `sum()`, `max()`, `len()`        |
| âš ï¸ **Con effetti collaterali** | Modifica lo stato o produce output               | `print()`, `append()`, `write()` |
